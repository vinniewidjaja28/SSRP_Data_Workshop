```python
data_americas_trimmed = data_americas.iloc[0:,1:]
data_americas_trimmed
```

```python
mask = data_americas_trimmed > 2000
data_americas_trimmed[mask]
```

```python
data_americas_trimmed[mask].describe()
```

```python
# mask: key to tell you which data sets you want specifically (like a filter)
# count: number of countries that had a GDP over 2000 
```

```python
mask_higher = data_americas_trimmed > data_americas_trimmed.mean()
data_americas_trimmed[mask_higher]
```

```python
data_americas_trimmed[mask_higher].describe()
```

```python
#GOAL: Compute a "wealth index"
#Define the wealth index to be the percentage of years for which a country's GDP is greater than the mean

mask_higher = data_americas_trimmed > data_americas_trimmed.mean()
wealth_score = mask_higher.aggregate('sum', axis=1) / len(data_americas_trimmed.columns) # lens gives you percentage

wealth_score
# count amount of trues in a row 
# ex. Argentina for 11 years has a GDP higher than the average 
```

```python
#EXCERCISE
#GOAL: Give the average life-expectancy grouped by column
#Step 1: Import Data
data_all = pd.read_csv('data/gapminder_all.csv', index_col = 'country')

#Step 2: Choose the columns that have life-expectancy 
data_all_trimmed = pd.concat([data_all.loc[:,'continent'],
                                         data_all.loc [:,'lifeExp_1952':'lifeExp_2007']],
                                         axis=1)
#Step 3: Group the data by continent and calculate the mean
mean_lifeexpectancy = data_all_trimmed.groupby('continent').mean()
mean_lifeexpectancy
```
